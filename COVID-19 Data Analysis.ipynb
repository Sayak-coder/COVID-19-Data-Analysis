```python
df6=df6.drop(['WHO Region'],axis=1)
```

```python
df6=pd.merge(df6,df1,on='Country/Region')
```

```python
df6=df6.drop(['Country/Region','NewCases','NewDeaths','TotalRecovered','NewRecovered','ActiveCases','Deaths/1M pop'],axis=1)
```

```python
df6.columns
```

```python
df6.isna().sum()
```

```python
df6['Serious,Critical']=df6['Serious,Critical'].fillna(0)
```

```python
df6=df6.dropna()
```

```python
df6.info()
```

```python
df6['WHO Region'].unique()
```

```python
df6['Continent'].unique()
```

```python
df6['WHO Region']=df6['WHO Region'].map({'Americas':0, 'South-East Asia':1, 'Europe':2, 'Africa':3,
                                         'Eastern Mediterranean':4, 'Western Pacific':5})
df6['Continent']=df6['Continent'].map({'South America':0, 'Asia':1, 'Europe':2, 'Africa':3,
                                        'North America':4, 'Australia/Oceania':5})
```

```python
cols_to_scale=df6.columns

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
df6[cols_to_scale]=scaler.fit_transform(df6[cols_to_scale])
```

```python
df6.describe()
```

```python
X=df6.drop('TotalDeaths',axis=1)
y=df6['TotalDeaths']

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=45)
```

```python
from sklearn.linear_model import LinearRegression
model_lr=LinearRegression()
model_lr.fit(X_train,y_train)
model_lr.score(X_test,y_test)
```
